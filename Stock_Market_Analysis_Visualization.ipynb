```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv_url="https://raw.githubusercontent.com/anshsaboo/nvidia-stock-forecast/refs/heads/data/nvidia_stock_cleaned.csv"
df=pd.read_csv(csv_url)
```

```python
df['date']=pd.to_datetime(df['date'])
df.set_index('date',inplace=True)
```

```python
sns.set(style="darkgrid")
```

```python
plt.figure(figsize=(10,10))
plt.plot(df.index,df['close'],label='Close Price',color='blue')
plt.title('NVIDIA Close Price Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.plot(df.index,df['volume'],label="Volume",color='red')
plt.title('NVIDIA Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
summary_stats= df.describe()
```

```python
print("The summary statistics are: \n",summary_stats)
```

```python
import seaborn as sns
sns.boxplot(data=df[['close', 'volume']])
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x='volume',y='close',data=df,alpha=0.5)
sns.regplot(x='volume',y='close',data=df,scatter=False,color='green')
plt.title('Close Price vs Volume')
plt.xlabel('Volume')
plt.ylabel('Close Price')
plt.tight_layout()
plt.show()

```
